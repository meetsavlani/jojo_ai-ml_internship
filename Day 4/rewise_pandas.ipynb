```python
def most_runs_against_each_team(batsman):
    mr = deliveries[deliveries['batsman']==batsman]
    print(mr.groupby('bowling_team')['batsman_runs'].sum().sort_values(ascending=False).head(3))
most_runs_against_each_team('V Kohli')
most_runs_against_each_team('RG Sharma')
```

```python
# most destructive batsman of 16-20 overs
# highest strikerate and minimum 100 balls faced in 16-20 overs

death=deliveries[deliveries['over']> 15]
all_batsmen_in_death=death.groupby('batsman')['batsman_runs'].count()# isse pata chlega kitni balls face kari hai
x = all_batsmen_in_death>200
all_batsmen_in_death[x].index 
batsmen_list = all_batsmen_in_death[x].index.tolist()

final = death[death['batsman'].isin(batsmen_list)]


runs = final.groupby('batsman')['batsman_runs'].sum()
balls = final.groupby('batsman')['batsman_runs'].count()

sr = (runs/balls)*100
sr
```

MERGE TWO DATAFRAMES

```python
# merging matches and deliveries dataset
new=deliveries.merge(data, left_on='match_id', right_on='id')
new.shape
```

```python
# new.groupby(['season','batsman'])['batsman_runs'].sum() # this is series ab isko dataframe me convert krna pdega jo reset_index() se hoga

# new.groupby(['season','batsman'])['batsman_runs'].sum().sort_values(ascending=False)

# new.groupby(['season','batsman'])['batsman_runs'].sum().sort_values(ascending=False).reset_index()

# ab humko most runs and season mil gayi hai ab simply hum duplicates hata skte hai season k basis pr
orange=new.groupby(['season','batsman'])['batsman_runs'].sum().sort_values(ascending=False).reset_index().drop_duplicates('season',keep='first')
# ye sb players mil gaye humko orange cap ke
# ab season wise sort kr skte hai for good visualization
orange.sort_values(by='season')
```

```python
# now i will try for purple
new.columns
# bowler , 

# new['player_dismissed'].head(20)
# new['dismissal_kind'].unique() # caught , bowled , lbw , caught and bowled , stumped , hit wicket 


dismissal_kind_bowler_wicket_values=new[new['dismissal_kind'].isin(['caught','bowled','lbw','caught and bowled','stumped','hit wicket'])]
bowler_wickets_count=dismissal_kind_bowler_wicket_values.groupby(['season','bowler'])['dismissal_kind'].count().sort_values(ascending=False)
dataframe_wickets=bowler_wickets_count.reset_index()
dataframe_wickets.drop_duplicates('season',keep='first').sort_values(by='season')
```
